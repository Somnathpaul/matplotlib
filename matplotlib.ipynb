```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas_df = pd.read_csv('gas_prices.csv')
gas_df
```

```python
plt.figure(figsize=(8,5))

plt.title('Gas Prices over time in $ [US]')

plt.plot(gas_df.Year, gas_df.UK, 'r.-', label='UK')
plt.plot(gas_df.Year, gas_df.USA, 'g.-', label='US')
plt.plot(gas_df.Year, gas_df['South Korea'], 'b.-', label='SK') # if column name has space

plt.xticks(gas_df.Year, rotation=40)

plt.xlabel('Year')
plt.ylabel('US $')

plt.legend()
plt.show()
```
